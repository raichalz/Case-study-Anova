3. Check whether all the zones generate the same amount of sales.

```python
# Performing ONE-WAY ANOVA USING SCIPY.STATS
```

```python
#null hypothesis 
#h0:all the zones generate the same amount of sales.
```

```python
#alternate hypothesis 
#h1:all the zones generate different amount of sales.
```

```python
f_oneway(data["Zone - A"],data["Zone - B"],data["Zone - C"],data["Zone - D"])
```

Here p valuve<0.05, So we can say that null hypothesis is not true and we reject it and conclude that there is zone wise difference in sales
